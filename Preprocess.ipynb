```python
import pandas
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
#import tensorflow as tf
#from keras.utils import to_categorical
import feather
import random
```

```python
FILENAME = "preproccesed_outliers_1085575_col_11_threshold_09"
```

```python
data = pd.read_csv("{}.csv".format(FILENAME))
data.head()
```

```python
def in_name(c):
    for col_name in ['Wind_Direction', 'Timezone', 'State']:
        if c.startswith(col_name):
            return True
    return False

remove = [col for col in data.columns if in_name(col)]
# remove
```

```python
# Filter the data
drop = ['Unnamed: 0', 'TMC', 'ID', 'Start_Time', 'End_Time', 'Weather_Timestamp', 'Description', 
        'City', 'County', 'Airport_Code', 
        'Weather_Condition',
        'Street', 'Turning_Loop', 'Bump', 'No_Exit', 'Roundabout', 'Give_Way', 'Traffic_Calming']
drop = drop+remove
data_filtered = data.drop(drop, axis=1).dropna().reset_index(drop=True)

# Convert to feather file format
data_filtered.to_feather("{}.feather".format(FILENAME))
```
