```python
import feather
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

FILENAME = "Datasets/US_Accidents_June20"
```

```python
#Putting data into feather
data = pd.read_csv(FILENAME+".csv")
data.to_feather("{}.feather".format(FILENAME))
```

```python
#Run this if the dataset for feather already produced

data = feather.read_dataframe("{}.feather".format(FILENAME))
```

## Severity and States

```python
data = data.drop(["ID","TMC","End_Lat","End_Lng"],axis = 1)
```

```python
data = data.dropna()
```

```python
data2 = data.groupby(['Severity','State']).size().reset_index().pivot(\
    columns='Severity', index='State', values=0)
data2.plot(kind='bar', stacked=True, title='Severity Count by States')
```
