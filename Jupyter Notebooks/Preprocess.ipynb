```python
data = data.dropna() #drop rows with missing values
```

```python
data['Start_Time'] = pd.to_datetime(data['Start_Time'])
data['End_Time'] = pd.to_datetime(data['End_Time'])

time = pd.DatetimeIndex(data['End_Time'] - data['Start_Time'])
data['Duration'] = time.hour * 3600 + time.minute*60 + time.second

# Remove instances where duration < 0
data = data[data['Duration']>=0]
```

```python
# Filter the data
drop = ['Unnamed: 0', 'ID', 'Start_Lat', 'Start_Lng', 'Start_Time', 'End_Time', 'Weather_Timestamp', 'Description', 
        'City', 'Airport_Code',
        'County',
        'Weather_Condition', 'Wind_Direction', 'Timezone', 'State', 
        'Street', 'Bump', 'No_Exit', 'Roundabout', 'Give_Way', 'Traffic_Calming']
# drop = drop+remove
data_filtered = data.drop(drop, axis=1).reset_index(drop=True) # Also remove rows with missing values

# Convert to feather file format
data_filtered.to_feather("{}.feather".format(FILENAME))
```
